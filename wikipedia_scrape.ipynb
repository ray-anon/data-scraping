```python
import requests
from bs4 import BeautifulSoup
def get_content_value(row_data):
    if row_data is None:
        return None
    if row_data.find("li"):
          return [li.get_text(" " , strip=True).replace("\xa0" , " ") for li in row_data.find_all("li")]
    else:
        return row_data.get_text(" " , strip=True).replace("\xa0" , " ")
def get_info_box(url):
    r = requests.get(url)
    contents = BeautifulSoup(r.content , 'lxml')
    info_box = contents.find(class_= "infobox vevent")
    movie_info = {}
    if info_box is None:
        info_box = contents.find(class_= "infobox vcard")
        if info_box is None:
          info_box = contents.find(class_= "infobox")
          if info_box is None:
            return
    info_rows = info_box.find_all('tr')
    for index ,row in enumerate(info_rows):
            if index == 0:
                continue
            elif index == 1:
                continue
            else:
                th_element = row.find("th")
                if th_element is None:
                    continue
                content_key = row.find("th").get_text(" " , strip=True)
                content_value = get_content_value(row.find("td"))
                movie_info[content_key] = content_value
    return movie_info

url = "https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films"
html_text = requests.get(url).content
soup = BeautifulSoup(html_text , 'lxml')
movies = soup.select(".wikitable.sortable i a")
base_path = "https://en.wikipedia.org"
movie_info_list = []
for index , movie in enumerate(movies):
    try:
        relative_path = movie['href']
        fullpath = base_path + relative_path
        title = movie['title']
        movie_info_list.append(get_info_box(fullpath))
    except Exception as e:
        print(e)
        print(movie.get_text())
```

```python
import json
def save_data(title , data):
  with open(title , 'w' , encoding="utf-8") as f:
     json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
def load_data(title):
  with open(title ,  encoding="utf-8") as f:
    return json.load(f)
```

```python
save_data('walt.json' , movie_info_list)
```

```python
load_data('walt.json')
```
